Data Visualization

```python
sns.distplot(data['total_bill'])
```

```python
sns.scatter(data['total_bill'].data['tip'])
plt.show()
```

```python
sns.scatterplot(x='total_bill',y='tip',data=data)
plt.show()
```

```python
sns.countplot(x="day",data=data.head(150),hue="smoker")
plt.show()
```

```python
sns.countplot(y="tip",data=data,hue="smoker")
plt.show()
```

```python
sns.violinplot(y="day",x="total_bill",data=data,hue="sex")
plt.show()
```

```python
sns.pairplot(data=data,hue="sex",palette="mako")
plt.show()
```

```python
sns.boxplot(x="day",y="tip",data=data)
plt.show()
```

```python
sns.heatmap(data.corr(),annot=True)
plt.figure(figsize=(5,5))
```
